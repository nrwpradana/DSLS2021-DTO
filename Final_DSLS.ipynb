```python
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
dat_cl=df2_Ehac[['personalDetail_travelPurpose','Age_Year','travelDetail_transportation_vehicleType','personalDetail_gender','personalDetail_citizen','travelDetail_destination_provinceName']]
```

```python
cat_var = ['personalDetail_travelPurpose','travelDetail_transportation_vehicleType','personalDetail_gender','personalDetail_citizen','travelDetail_destination_provinceName']
for var in cat_var:
    cat_list = pd.get_dummies(dat_cl[var], drop_first=True, prefix=var)
    data1 = dat_cl.join(cat_list)
    dat_cl = data1
dat_cl.head()
```

```python
dat_cl=dat_cl.drop(columns=['personalDetail_travelPurpose','travelDetail_transportation_vehicleType','personalDetail_gender','personalDetail_citizen','travelDetail_destination_provinceName'])
dat_cl.head()
```

```python
x_array = np.array(dat_cl)
print(x_array)
```

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_array)
x_scaled
```

```python
dbi = []
K = range(2,10)
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(x_scaled)
    labels=km.labels_
    dbi.append(davies_bouldin_score(x_scaled, labels))
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 10), dbi)
plt.xticks(range(2, 10))
plt.xlabel("Number of Clusters")
plt.ylabel("DBI")
plt.show()
```
